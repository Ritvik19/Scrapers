# Scraping URLs to be scrapped

## Language URLs

```python
language_dict = {}
url = 'https://syntaxdb.com/reference'
res = requests.get(url)
if res.status_code == requests.codes.ok:
    ressoup = bs4.BeautifulSoup(res.text, 'lxml')
    elems = ressoup.select('#language-dropdown .deep-purple-text')
    for e in elems:
        language_dict[e.getText().strip()] = 'https://syntaxdb.com'+str(e.get('href'))
else:
    print('Something went wrong')
for k, v in language_dict.items():
    print(k, v)
```

```python
urls = []
res = requests.get('https://syntaxdb.com/ref/java')
if res.status_code == requests.codes.ok:
    ressoup = bs4.BeautifulSoup(res.text, 'lxml')
    elems = ressoup.select('ul li a')
    for i, e in enumerate(elems[10:-50]):
#         print(i)
#         print(e.getText())
#         print(e.get('href'))
        if e.get('href'):
            urls.append('https://syntaxdb.com'+str(e.get('href')))
else:
    print('Something went wrong')
syntax_dict['Java'] = urls   
print(syntax_dict['Java'], len(syntax_dict['Java']))
```

# Scraping

```python
counter = 1
syntax_db = pd.DataFrame(columns=['Language' ,'Title', 'Code', 'Notes', 'Example'])
for k, val in syntax_dict.items():
    for v in val:
        clear_output(wait=True)
        print(counter,'/269')
        counter+=1
        res = requests.get(v)
        if res.status_code == requests.codes.ok:
            ressoup = bs4.BeautifulSoup(res.text, 'lxml')
            notes = ''
            elems = ressoup.select('.col.s12.m8.l8 p')
            for i, e in enumerate(elems):
                notes += e.getText()+'\n'
            new_syntax = pd.DataFrame({
                'Language': [k], 'Title': [ressoup.select('h5')[0].getText()], 'Notes': [notes],
                 'Code':[ressoup.select('code')[0].getText()], 'Example':[ressoup.select('code')[1].getText()]
            })
            syntax_db = pd.concat([syntax_db, new_syntax], axis=0)
        else:
            print('Something went wrong')
```

```python
syntax_db = syntax_db.reset_index(drop=True)
```
